```python
import json
from pathlib import Path

manifest = json.loads(Path("cache/degradation_manifest.json").read_text())

jpeg_qualities = [
    entry["params"]["jpeg_quality"]
    for entry in manifest["entries"]
    if entry.get("params") and entry["params"]["jpeg_quality"] is not None
]

no_blur_count = sum(
    1 for entry in manifest["entries"]
    if entry.get("params") and entry["params"]["blur_type"] == "none"
)

no_noise_count = sum(
    1 for entry in manifest["entries"]
    if entry.get("params") and entry["params"]["noise_type"] == "none"
)

total = len(manifest["entries"])
avg_jpeg = sum(jpeg_qualities) / len(jpeg_qualities)

print(f"Avg JPEG quality:      {avg_jpeg:.1f}  (want below 65)")
print(f"No blur applied:       {no_blur_count/total*100:.1f}%  (want below 25%)")
print(f"No noise applied:      {no_noise_count/total*100:.1f}%  (want below 35%)")
```
